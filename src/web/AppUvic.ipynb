```python
import pandas as pd
# import sqlite3
```

```python
pd.options.plotting.backend = "plotly"
```

```python
cols_select = ['ID', 
               'titulo',
               'fecha_publicacion',
               'url',
               'cluster']

# cnx = sqlite3.connect('data.db')
# df = pd.read_sql_query("SELECT * FROM news", cnx)
df = pd.read_csv('../data/news_categorized.csv',
              usecols = cols_select)
```

```python
# df.info()
```

```python
df['fecha_publicacion']=pd.to_datetime(df['fecha_publicacion'])
```

```python
df['year']=df['fecha_publicacion'].dt.year
df['week']=df['fecha_publicacion'].dt.week
df['week'] = ['0'+ str(x) if len(str(x)) == 1 else str(x) for x in df['week']]

df['year_week']=df['year'].astype(str) + ' ' +df['week']
```

```python
df.head()
```

```python
df_conteo = (
    df
    .groupby(['cluster', 'year_week', 'year', 'week'])
    .size()
    .reset_index()
)
```

```python
df_conteo.rename(columns = {0 : 'num_hechos'}, inplace = True)
```

```python
df_conteo['year_week'] = df_conteo['year_week'].astype(str)
```

```python
import plotly.express as px
```

```python
from datetime import datetime
```

```python
df_conteo['date_year_week'] = [datetime.strptime(x + ' 0', "%Y %W %w") for x in df_conteo['year_week']]
```

```python
df_conteo['year_month'] = df_conteo['year'].astype(str) + '-' + df_conteo['date_year_week'].dt.month.astype(str)
```

```python
df_conteo.to_csv('df_conteo.csv', index = False)
```

```python
df_conteo2 = df_conteo.groupby(['cluster', 'year_month']).agg({'num_hechos': 'sum'})
```
